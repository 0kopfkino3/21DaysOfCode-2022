```python
n = 5
for i in range(n,0,-1):
  for j in range(n,0,-1):
    if(j>i):
      print(j, end= "   ")
  for j in range(1,2*i):
    print(i, end= "   ")
  for j in range(i+1,n+1):
    print(j, end= "   ")
  print("\n")
for i in range(1,n):
  for j in range(n,i,-1):
    if(j>i):
      print(j, end= "   ")
  for j in range(1,2*i):
    print(i+1, end= "   ")
  for j in range(i+1,n+1):
     print(j, end= "   ")
  print("\n")
  
```
